```python
import pandas as pd
import numpy as np
```

```python
X_train = pd.read_csv("training_set_features.csv")
X_test = pd.read_csv("test_set_features.csv")
y_train = pd.read_csv("training_set_labels.csv")
```

```python
X_train = X_train.drop(['employment_occupation', 'employment_industry', 'hhs_geo_region'],axis='columns')
cols=[ 'age_group', 'education', 'race', 'sex', 'income_poverty', 'marital_status', 'rent_or_own', 'employment_status', 'census_msa']
X_train = pd.get_dummies(X_train, columns=cols)
cols=['xyz_concern', 'xyz_knowledge',
       'behavioral_antiviral_meds', 'behavioral_avoidance',
       'behavioral_face_mask', 'behavioral_wash_hands',
       'behavioral_large_gatherings', 'behavioral_outside_home',
       'behavioral_touch_face', 'doctor_recc_xyz', 'doctor_recc_seasonal',
       'chronic_med_condition', 'child_under_6_months', 'health_worker',
       'health_insurance', 'opinion_xyz_vacc_effective', 'opinion_xyz_risk',
       'opinion_xyz_sick_from_vacc', 'opinion_seas_vacc_effective',
       'opinion_seas_risk', 'opinion_seas_sick_from_vacc', 'household_adults',
       'household_children']
for col in cols:
    X_train[col].fillna(X_train[col].median(), inplace = True)
X_train
```

```python
X_test = X_test.drop(['employment_occupation', 'employment_industry', 'hhs_geo_region'],axis='columns')
cols=[ 'age_group', 'education', 'race', 'sex', 'income_poverty', 'marital_status', 'rent_or_own', 'employment_status', 'census_msa']
X_test = pd.get_dummies(X_test, columns=cols)
cols=['xyz_concern', 'xyz_knowledge',
       'behavioral_antiviral_meds', 'behavioral_avoidance',
       'behavioral_face_mask', 'behavioral_wash_hands',
       'behavioral_large_gatherings', 'behavioral_outside_home',
       'behavioral_touch_face', 'doctor_recc_xyz', 'doctor_recc_seasonal',
       'chronic_med_condition', 'child_under_6_months', 'health_worker',
       'health_insurance', 'opinion_xyz_vacc_effective', 'opinion_xyz_risk',
       'opinion_xyz_sick_from_vacc', 'opinion_seas_vacc_effective',
       'opinion_seas_risk', 'opinion_seas_sick_from_vacc', 'household_adults',
       'household_children']
for col in cols:
    X_test[col].fillna(X_test[col].median(), inplace = True)
X_test
```
